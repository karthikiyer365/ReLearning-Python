## **Data Structures**
Data Structures are different ways that we can store the data within our program. The Structures python uses are:


### Lists


- `list = []`  : is a mutable sequences of elements


```python
# Lists
fruits = ["apple", "banana", "orange", "grape", "pineapple", "mango"]
fruits
```

```python
# Accessing elements in a list
print("First fruit:", fruits[0])
print("Last fruit:", fruits[-1])
```

```python
# Adding items
fruits.append("kiwi")
fruits
```

```python
# Remove items
fruits.remove("pineapple")
fruits
```

```python
# Accessing tuple elements
print("X coordinate:", coordinates[0])
print("Y coordinate:", coordinates[1])
```

### Dictionary

- `dictionary = {}` : is an unordered collections of key-value pairs



```python
# Example: Working with dictionaries
student = {
    "name": "John",
    "age": 20,
    "grade": "A",
    "subjects": ["Math", "Science", "English"],
    "Zipcode": 22004
}
```

```python
# Accessing dictionary elements
print("Student name:", student["name"])
print("Student age:", student["age"])
```

```python
# Modifying dictionary
student["age"] = 21
print("Updated student age:", student["age"])
```

```python
del student["Zipcode"]
student
```

## **Conditional Statements**

Conditional statements in Python allow you to execute different blocks of code based on whether a condition is true or false.

The three important parts of conditional statements
- `if` (to check primary condition)
- `elif` (to check condition in addition to `if`)
- `else` (in case no other condition is true)

```python
num = int(input('Enter a number ->  '))

if num > 0:
    print("Positive")
elif num == 0:
    print("Equal to zero")
else:
    print("Negative")
```

## **Loops**
Loops are used in programming to execute a block of code repeatedly until a certain condition is met


There are 2 ways of looping in python
- `for`
- `while`

### For loops
- for loop is used to iterate over a sequence (such as a list, tuple, string, or range) or other iterable objects.



```python
fruits = ["apple", "banana", "cherry"]

print('\n For fruits list')
for fruit in fruits:
    print(fruit)
```

```python
fruits2 = ("mango", "strawberry", "orange")

print('\n For fruits2 list')
for item in fruits2:
    print(item)
```

```python
one_fruit = "watermelon"

print('\n For one_fruit list')
for i in one_fruit:
    print(i)
```

You can also use for loops over a range.
Consider following example

```python
# To go from 0 to 10
for idk in range(0,10):
    print(idk)

'''
Note: the loop doesn't run when idk becomes the last value
'''
```

```python
# To go from 0 to 30 """in multiples of 3"""
for idk in range(0,30,3):
    print(idk)
```

### While loop
- A while loop is used to repeatedly execute a block of code as long as a specified condition is true.

```python
i = 0
while i <= 5:
    print(f'i is {i} hence loop runs')
    i += 1
```

### Loop control statements
- `continue`: skips current iteration
- `break`: stops loop

```python
for i in range(0,20,3):
    if i%5 == 0:
        continue
    elif i == 18:
        break
    else:
        pass
    print(i)
```
